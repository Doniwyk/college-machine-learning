### 1. Import Library

```python
import pandas as pd
from sklearn.model_selection import train_test_split, KFold
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

### 2. Load Data

```python
data = './data/wbc.csv'
df = pd.read_csv(data)
df.head()
```

```python
df.info()
df.isnull().sum()
```

### 3. Slice Data
> id and unnamed:32 will not be used because its not an attribute that influences cancer classification.

```python
df = df.iloc[:, 1:-1]
df.head()
```

### 4. Encoding

```python
encoder = OneHotEncoder(sparse_output=False)

diagnosis_encoded = encoder.fit_transform(df[['diagnosis']])
diagnosis_encoded_df = pd.DataFrame(diagnosis_encoded, columns=encoder.get_feature_names_out(['diagnosis']))
df_encoded = pd.concat([df, diagnosis_encoded_df], axis=1)
df_encoded = df_encoded.drop(columns=['diagnosis'])

df_encoded.head()
```

### 5. Standarization
> Standarization exclude the diagnosis_B and diagnosis_M using differentiation because its a label data 

```python
std = StandardScaler()

numeric_columns = df_encoded.columns.difference(['diagnosis_B', 'diagnosis_M'])
df_encoded[numeric_columns] = std.fit_transform(df_encoded[numeric_columns])

df_encoded.head()
```

### 6. Stratified Split Data

```python
x = df_encoded.drop(columns=['diagnosis_B', 'diagnosis_M'])
y = df_encoded[['diagnosis_B', 'diagnosis_M']]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=0)

print(len(x_test))
print(len(x_train))
print(len(y_test))
print(len(y_train))
```
